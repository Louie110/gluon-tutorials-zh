# matplotlib  
 ## [官方文档](https://matplotlib.org/)

````python
import matplotlib.pyplot as plt  #导入绘图库
import numpy as np   #导入数组库
import matplotlib as mpl  #为了在图文中显示中文，所以导入matplotlib
````

````python
#线内画面
%matplotlib inline 

#GUI画面
#%matplotlib tk      

#解决中文显示问题
mpl.rcParams['font.sans-serif'] = ['FangSong']  #这里设置中文字体为仿宋字体
mpl.rcParams['axes.unicode_minus'] = False
````

````python
x = np.arange(0,2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8,10))  #建立figure对象，并设置图的显示大小

plt.plot(x,y1,'g*-',label = 'sin(x)',linewidth = 4)
plt.plot(x,y2,'ro--',label = 'cos(x)',alpha=0.7)      #设置透明度为0.7
plt.axis([0,2*np.pi,0,1])  
plt.xlabel('x取值')
plt.ylabel('y取值')
plt.title('matplotlib示例1')
plt.grid(True)   #显示虚线框
plt.legend()    #显示图示说明
plt.savefig('test1.png',dpi = 40)   #将当前的figure保存为图像文件，设置图像分辨率为40，则输出图像的宽度为8*40=320个像素
plt.show()
````

# 面向对象式画图

````python
fig = plt.figure(figsize = (20,10)) #建立对象fig

#指定每个子图在图形区域内的位置
ax1 = fig.add_subplot(2,2,1)  
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,1,2)

ax3.plot(np.random.randn(50).cumsum(),'y-',alpha = 0.8)  #在当前axes上绘制曲线(ax3)
ax1.hist(np.random.randn(50),bins = 20, color= 'g',alpha = 0.5)  #在ax1上绘制直方图
ax2.scatter(np.arange(30),np.arange(30) + 3*np.random.randn(30),color = 'r')  #在ax2上绘制散点图
plt.show()  #展示
````

# 多子图  
- 一个Figure对象可以包含多个子图，在matplotlib中用Axes对象表示一个绘图区域。调用方式为： 
```py
subplot(numRows,numCols,plotNum)
```  
整个绘图区域被等分为numRows行和numCols列，然后按照从左到右，从上到下的顺序对每个区域进行编号，左上区域的编号为1，plotNum参数指定 所创建Axes对象的区域。若numRows、numCols、plotNum三个参数都小于10，则可以把它们缩写成一个整数，例如subplot(323)和subplot(3,2,3)的含义相同。如果新建的子图和之前所创建的子图有重叠的部分，之前的子图将被删除。

````python
fig,axes = plt.subplots(2,2,sharex = True, sharey = True)  #sharex表示x轴坐标共享，sharey表示y轴坐标共享
#print(fig)
#print(axes)
````

````python
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins = 10,color = 'red',alpha = 0.5)

plt.subplots_adjust(wspace=0,hspace=0) #调整各图形间的距离
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,1000) 
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(10,4))
plt.plot(x,y,'r*-',label = 'sin(x)',linewidth = 2)
plt.plot(x,z,'b<--',label = 'cos(x)',linewidth = 3)
plt.axis([0,11,-2,2])   #设置x轴的范围和y轴的范围

##    使用annotate 函数进行注释，annotate 主要有两个参数：
#      xy ：注释位置 
#      xytext ：注释文字位置
#      shrink : 改变箭头长短，值越小，箭头越长；值越大，箭头越小
plt.annotate('sin(x)',xy=(3, 0.3), xytext=(3.8, 0.6),arrowprops=dict(facecolor='m', shrink=0.05))
plt.annotate('cos(x)', xy=(8, -0.2), xytext=(7, -0.8),arrowprops=dict(facecolor='c', shrink=0.05))

#添加标题
plt.xlabel('横轴')
plt.ylabel('纵轴')
plt.title('matplotlib 测试2',fontsize=18)  #fontsize改变所设置的title字体大小

#设置边框
ax = plt.gca()  #获取坐标轴信息
ax.spines['top'].set_color('none')   #设置顶端没有边框
ax.spines['right'].set_color('green')#设置右边框为绿色
ax.spines['bottom'].set_color('red')  #设置下边框为红色
ax.spines['left'].set_color('yellow') #设置左边框为黄色

plt.ylim(-1.2,1.2)
plt.legend()
plt.grid(True) #显示虚线框
plt.savefig('2.png',dpi=80)
plt.show()
````

# 柱状图(条形图)

````python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 设置在jupyter中matplotlib的显示情况
%matplotlib inline

# 解决中文显示问题
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10,8))
plt.bar([1,3,5,7,9,11],[5,2,7,8,2,6],align = 'center',label="Example one",color = 'y',edgecolor = 'r')

plt.bar([2,4,6,8,10,12],[8,6,2,5,6,3], label="Example two", color='g')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')

#设置x轴刻度标签
plt.xticks(np.arange(1,13),[str(i)+'月' for i in range(1,13)])

plt.title('柱状图')

plt.show()
````

# 练习

````python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
# 设置在jupyter中matplotlib的显示情况
%matplotlib inline

# 解决中文显示问题
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False

x = [1,2,3]
y1 = [600,700,800]
y2 = [500,900,1000]
plt.figure(figsize=(8,5))
plt.bar(x,y1,0.1,align = 'edge',label="进口 单位/美元",color = 'y')   #0.3表示柱状图条框的宽度，值为正的话表示向右取宽度
plt.bar(x,y2,-0.1,align = 'edge' ,label="出口 单位/美元", color='r')  #-0.3表示柱状图条框的宽度，值为负的话表示向左取宽度

#给柱状图添加说明
for m,n in zip(x,y1):
    plt.text(m+0.08, n+10, '%.2f' %n, ha='center',va= 'bottom')

for m,n in zip(x,y2):
    plt.text(m-0.08, n+10, '%.2f' %n, ha='center', va= 'bottom')
    
plt.legend()
plt.xlabel('第一季度/月')
plt.ylabel('进出口额/美元')

#设置x轴刻度标签
plt.xticks(np.arange(1,4),[str(i)+'月' for i in range(1,4)])

plt.title('第一季度进出口数据',fontsize=20)  #在这里fontsize和size一个意思，任选其一

plt.show()
````

# 对数坐标图  
绘制对数坐标图的函数有三个：semilogx()、semilogy()、loglog()  
分别绘制X轴为对数坐标、Y轴为对数坐标以及两个轴都为对数坐标的图表

````python
x = np.linspace(0.1,1000,1000)
y = np.abs(1/(1+0.1j*x))  #计算低通滤波器的频率响应

fig,axes = plt.subplots(2,2)
functions = ('plot','semilogx','semilogy','loglog')

for ax,fname in zip(axes.ravel(),functions):
    func = getattr(ax,fname)
    func(x,y,linewidth=2)
    ax.set_ylim(0,1.5)
````

# 直方图

````python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 设置在jupyter中matplotlib的显示情况
%matplotlib inline

# 解决中文显示问题
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False

#展现出10000个随机数满足 平均值是100 标准差是15 呈现的一个正态分布曲线
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
plt.figure(figsize=(10,8))

plt.hist(x, 100, normed=1,color='c', alpha=0.75)

plt.title('直方图',size=20)
plt.text(60, 0.025,r'$\mu=100,\sigma=15$')   # 添加文本
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
````

# 散点图

````python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

t=['r','y','c','b','g','m','b','r']  #指定的每个点的颜色序列

plt.figure()
plt.scatter(x,y, label='散点分布', color=t,s=25, marker="*",alpha = 0.8)

plt.xlabel('x')
plt.ylabel('y')
plt.title('散点图',size=20)
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(8,6))
n=1000
#rand 均匀分布和 randn高斯分布
x=np.random.randn(1,n)
y=np.random.randn(1,n)
T=np.arctan2(x,y) #np.random.rand(n) * 125

plt.scatter(x,y,c=T,s=30,alpha=0.8,marker='o')
#T:散点的颜色
#s：散点的大小
#alpha:是透明程度
plt.show()
````
